## Understanding the Evaluation Metric

Evaluation metrics are used to measure the quality of the statistical or machine learning model.There are many different types of evaluation metrics available to test a model. These include classification accuracy, logarithmic loss etc. For this particluar problem, our submissions will be evaluated using **F1** between the predicted and expected answers. 

The **F score**, also called the **F1 score** or **F measure**, is a measure of a test’s accuracy. 

The F score is defined as the weighted harmonic mean of the test’s precision and recall. 

![](https://imgur.com/nC4QwrO.png)

- Precision, also called the positive predictive value, is the proportion of positive results that truly are positive. 

- Recall, also called sensitivity, is the ability of a test to correctly identify positive results to get the true positive rate. 

![](https://upload.wikimedia.org/wikipedia/commons/thumb/2/26/Precisionrecall.svg/350px-Precisionrecall.svg.png)
*source:https://en.wikipedia.org/wiki/Precision_and_recall*

The  F score reaches the best value, meaning perfect precision and recall, at a value of 1. The worst F score, which means lowest precision and lowest recall, would be a value of 0. 



### Why is this Useful?
The F score is used to measure a test’s accuracy, and it balances the use of precision and recall to do it. The F score can provide a more realistic measure of a test’s performance by using both precision and recall. The F score is often used in information retrieval for measuring search, document classification, and query classification performance. 
          
          
source: https://deepai.org/machine-learning-glossary-and-terms/f-score

## Понимание метрики оценки

Метрика оценки используется для измерения качества статистической модели или модели машинного обучения.Cуществует множество различных типов оценочных метрик, доступных для тестирования модели. К ним относятся точность классификации, логарифмические потери и т. д. 
Для этой конкретной проблемы наши заявки будут оцениваться с использованием **F1** между прогнозируемым и ожидаемым ответами. 

**Оценка F**, также называемая **оценка F1** или **мера F**, является мерой точности теста.. 

Оценка F определяется как средневзвешенное гармоническое значение точности и запоминания теста. 

![](https://imgur.com/nC4QwrO.png)

- Точность(precision), также называемая положительной прогностической ценностью, - это доля действительно положительных результатов 

- Отзыв(recall), также называемый чувствительностью, - это способность теста правильно определять положительные результаты для получения истинно положительных результатов. 

![](https://upload.wikimedia.org/wikipedia/commons/thumb/2/26/Precisionrecall.svg/350px-Precisionrecall.svg.png)
*источник:https://en.wikipedia.org/wiki/Precision_and_recall*

Оценка F достигает наилучшего значения, означающего идеальную точность и отзывчивость, при значении 1. Наихудшая оценка F, которая означает самую низкую точность и самый низкий уровень отзыва, будет иметь значение 0. 



### Почему это полезно?
Оценка F используется для измерения точности теста и уравновешивает использование точности и отзывчивости при ее выполнении. Оценка F может обеспечить более реалистичную оценку эффективности теста за счет как точности, так и запоминания. Оценка F часто используется при поиске информации для измерения эффективности поиска, классификации документов и классификации запросов.

## 1. Data Cleaning

Before we start with any NLP project we need to pre-process the data to get it all in a consistent format.We need to clean, tokenize and convert our data into a matrix. Some of the  basic text pre-processing techniques includes:

* Make text all **lower case** or **uppercase** so that the algorithm does not treat the same words in different cases as different
* **Removing Noise** i.e everything that isn’t in a standard number or letter i.e Punctuation, Numerical values,  common non-sensical text (/n)
* **Tokenization**: Tokenization is just the term used to describe the process of converting the normal text strings into a list of tokens i.e words that we actually want. Sentence tokenizer can be used to find the list of sentences and Word tokenizer can be used to find the list of words in strings.
* **Stopword Removal**: Sometimes, some extremely common words which would appear to be of little value in helping select documents matching a user need are excluded from the vocabulary entirely. These words are called stop words

### More data cleaning steps after tokenization:

* **Stemming**: Stemming is the process of reducing inflected (or sometimes derived) words to their stem, base or root form — generally a written word form. Example if we were to stem the following words: “Stems”, “Stemming”, “Stemmed”, “and Stemtization”, the result would be a single word “stem”.
* **Lemmatization**: A slight variant of stemming is lemmatization. The major difference between these is, that, stemming can often create non-existent words, whereas lemmas are actual words. So, your root stem, meaning the word you end up with, is not something you can just look up in a dictionary, but you can look up a lemma. Examples of Lemmatization are that “run” is a base form for words like “running” or “ran” or that the word “better” and “good” are in the same lemma so they are considered the same.
* Parts of speech tagging
* Create bi-grams or tri-grams
And more...

However, it is not necessary that you would need to use all these steps. The usage depends on your problem at hand. ## 1. Очистка данных

Прежде чем мы начнем с любого проекта NLP, нам необходимо предварительно обработать данные, чтобы получить все в согласованном формате. Нам нужно очистить, токенизировать и преобразовать наши данные в матрицу. Некоторые из основных методов предварительной обработки текста включают:

* Сделайте текст полностью **нижним регистром** или **прописным**, чтобы алгоритм не рассматривал одни и те же слова в разных регистрах как разные
* **Удаление шума** то есть все, что не является стандартным числом или буквой, например пунктуация, числовые значения, обычный бессмысленный текст (/ n)
* **Токенизация**: Токенизация - это просто термин, используемый для описания процесса преобразования обычных текстовых строк в список токенов, то есть слов, которые нам действительно нужны. Токенизатор предложений можно использовать для поиска списка предложений, а токенизатор Word можно использовать для поиска списка слов в строках.
* **Удаление стоп-слов**: Иногда некоторые чрезвычайно распространенные слова, которые, казалось бы, не имеют большого значения при выборе документов, соответствующих потребностям пользователя, полностью исключаются из словаря. Эти слова называются стоп-словами

### Дополнительные шаги по очистке данных после токенизации:

* **Стемминг**: Стемминг - это процесс сокращения изменяемых (или иногда производных) слов до их основы, основы или корневой формы - обычно письменной словоформы. Например, если бы мы должны были выделить следующие слова: «Стебли», «Ствол», «Ствол», «и Стемтизация», результатом было бы одно слово «стержень».

* **Лемматизация**: небольшой вариант стемминга - лемматизация. Основное различие между ними состоит в том, что при выделении корней часто могут образовываться несуществующие слова, тогда как леммы являются реальными словами. Итак, корневая основа, означающая слово, которое вы получите, - это не то, что вы можете просто найти в словаре, вы можете найти лемму. Примерами лемматизации являются то, что «бег» является базовой формой для таких слов, как «бег» или «бег», или что слова «лучше» и «хорошо» находятся в одной лемме, поэтому они считаются одинаковыми.

````python
# A quick glance over the existing data Быстрый взгляд на существующие данные
train['text'][:5]
````

````python
# Applying a first round of text cleaning techniques Применение методов очистки текста в первом раунде

def clean_text(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers. Сделать текст строчными, удалить текст в квадратных скобках, удалить ссылки, удалить знаки препинания
    и удалите слова, содержащие числа.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

# Applying the cleaning function to both test and training datasets Применение функции очистки как к тестовым, так и к обучающим наборам данных
train['text'] = train['text'].apply(lambda x: clean_text(x))
test['text'] = test['text'].apply(lambda x: clean_text(x))

# Let's take a look at the updated text Посмотрим на обновленный текст
train['text'].head()
````

## 3. Stopwords Removal

Now, let's get rid of the stopwords i.e words which occur very frequently but have no possible value like **a, an, the, are **etc.

## 3. Удаление стоп-слов

Теперь давайте избавимся от стоп-слов, то есть слов, которые встречаются очень часто, но не имеют возможных значений, таких как  **a, an, the, are **etc.

````python
def remove_stopwords(text):
    """
    Removing stopwords belonging to english language Удаление игнорируемых слов, принадлежащих к английскому языку
    
    """
    words = [w for w in text if w not in stopwords.words('english')]
    return words


train['text'] = train['text'].apply(lambda x : remove_stopwords(x))
test['text'] = test['text'].apply(lambda x : remove_stopwords(x))
train.head()
````

## 4. Token normalization

Token normalisation means converting different tokens to their base forms. This can be done either by:

- **Stemming** :  removing and replacing suffixes to get to the root form of the word, which is called the **stem** for instance cats - cat, wolves - wolv 
- **Lemmatization** : Returns the base or dictionary form of a word, which is known as the **lemma** 

[*source*](https://www.coursera.org/learn/language-processing/lecture/SCd4G/text-preprocessing)

## 4. Нормализация токена

Нормализация токенов означает преобразование разных токенов в их базовые формы. Это можно сделать:

- **Стемминг**: удаление и замена суффиксов для перехода к корневой форме слова, которая называется ** основой **, например, cats - cat, wolves - wolv. 
- **Лемматизация**: возвращает базовую или словарную форму слова, известную как **лемма**.

[*источник*](https://www.coursera.org/learn/language-processing/lecture/SCd4G/text-preprocessing)

````python
# Stemming and Lemmatization examples Примеры стемминга и лемматизации
text = "feet cats wolves talked"

tokenizer = nltk.tokenize.TreebankWordTokenizer()
tokens = tokenizer.tokenize(text)

# Stemmer Стеммер
stemmer = nltk.stem.PorterStemmer()
print("Stemming the sentence: ", " ".join(stemmer.stem(token) for token in tokens))

# Lemmatizer Лемматизатор
lemmatizer=nltk.stem.WordNetLemmatizer()
print("Lemmatizing the sentence: ", " ".join(lemmatizer.lemmatize(token) for token in tokens))
````

It is important to note here that stemming and lemmatization sometimes donot necessarily improve results as at times we donot want to trim words but rather preserve their original form. Hence their usage actually differs from problem to problem. For this problem, I will not use these techniques.

Здесь важно отметить, что стемминг и лемматизация иногда не обязательно улучшают результаты, поскольку порой мы хотим не обрезать слова, а, скорее, сохранить их первоначальную форму. Следовательно, их использование на самом деле отличается от проблемы к проблеме. Для этой задачи я не буду использовать эти методы.

### Getting it all together- A Text Preprocessing Function
This concludes the pre-processing part. It will be prudent to convert all the steps undertaken into a function for better reusability.

### Собираем все вместе - функция предварительной обработки текста
На этом завершается часть предварительной обработки. Будет разумно преобразовать все предпринятые шаги в функцию для лучшего повторного использования.

````python
# text preprocessing function функция предварительной обработки текста
def text_preprocessing(text):
    """
    Cleaning and parsing the text. Очистка и парсинг текста.

    """
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
    
    nopunc = clean_text(text)
    tokenized_text = tokenizer.tokenize(nopunc)
    remove_stopwords = [w for w in tokenized_text if w not in stopwords.words('english')]
    combined_text = ' '.join(remove_stopwords)
    return combined_text
````

## Мешок слов

### TFIDF Features

A problem with the Bag of Words approach is that highly frequent words start to dominate in the document (e.g. larger score), but may not contain as much “informational content”. Also, it will give more weight to longer documents than shorter documents.

One approach is to rescale the frequency of words by how often they appear in all documents so that the scores for frequent words like “the” that are also frequent across all documents are penalized. This approach to scoring is called Term Frequency-Inverse Document Frequency, or TF-IDF for short, where:

**Term Frequency: is a scoring of the frequency of the word in the current document.**

```
TF = (Number of times term t appears in a document)/(Number of terms in the document)
```

**Inverse Document Frequency: is a scoring of how rare the word is across documents.**

```
IDF = 1+log(N/n), where, N is the number of documents and n is the number of documents a term t has appeared in.
```

### Возможности TFIDF

Проблема с подходом «Мешок слов» заключается в том, что в документе начинают преобладать очень часто встречающиеся слова (например, с большим количеством очков), но они могут не содержать столько «информационного содержания». Кроме того, это придаст больший вес более длинным документам, чем более коротким.

Один из подходов состоит в том, чтобы изменить частоту слов в зависимости от того, как часто они появляются во всех документах, так что оценки за часто встречающиеся слова, такие как «the», которые также часто встречаются во всех документах, были оштрафованы. Такой подход к оценке называется Term Frequency-Inverse Document Frequency или сокращенно TF-IDF, где:

**Term Frequency: оценка частоты встречаемости слова в текущем документе.**

```
TF = (Количество раз, когда термин t встречается в документе) / (Количество терминов в документе)
```

**Частота обратного документа: оценка того, насколько редко слово встречается в документах.**

```
IDF = 1+log(N/n), где, N - количество документов, а n - количество документов, в которых появился термин t..
```
